```python
corpus = []
for i in range(len(dreams)):
    review = re.sub('[^a-zA-Z]', ' ', dreams[i])
    review = review.lower()
    review = review.split()
    ps = ItalianStemmer()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('italian'))]
    review = [x for x in review if not x.startswith('sogn')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
# Creating the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1500)
X = cv.fit_transform(corpus).toarray()
```

```python
X[1]
```

# Hierarchical clustering

```python
import scipy.cluster.hierarchy as sch
```

```python
dendrogram = sch.linkage(X, method = 'ward')
```

```python
dendrogram_plot = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Frasi')
plt.ylabel('Euclidean distances')
plt.show()
plt.gcf()
plt.savefig("histo2.pdf")
```

```python
plt.figure(figsize=(150,150))
dendro  = sch.dendrogram(dendrogram, leaf_rotation=90.)
plt.rcParams.update({'font.size': 100})
plt.title('Dendrogram',)
plt.xlabel('Frasi')
plt.ylabel('Euclidean distances')
#plt.savefig('plt.png', format='png', bbox_inches='tight')
plt.savefig('plt.jpg', format='jpg', bbox_inches='tight')
```

```python
th = 40
cluster_labels = sch.fcluster(Z = dendrogram, t = th, criterion= 'distance')
```
